```python
import os
import pandas as pd

import pyarrow.parquet as pq
import pyarrow as pa
```

```python
country_code = 'SK'
save_path = f'data/places/{country_code}.parquet'
base_dir = 'data/places/type=place'
pqwriter = None

for filename in os.listdir(base_dir):

    print(f'Processing {filename}')

    fp = os.path.join(base_dir, filename)

    # Open the Parquet file
    parquet_file = pq.ParquetFile(fp)

    if not pqwriter:
        pqwriter = pq.ParquetWriter(save_path, parquet_file.schema.to_arrow_schema())

    n_processed = 0
    batch_size = 10000

    # Process the file in chunks
    for batch in parquet_file.iter_batches(batch_size=batch_size):

        print(f'Processing chunk {n_processed} - {n_processed + batch_size}')

        # Convert the RecordBatch to a Table
        chunk_table = pa.Table.from_batches([batch])
        chunk_table = chunk_table.to_pandas()

        country_codes = chunk_table['addresses'].apply(lambda x: None if x is None else x[0].get('country'))
        filtered_df = chunk_table[country_codes == country_code]
        chunk_table_filtered = pa.Table.from_pandas(filtered_df, schema=parquet_file.schema.to_arrow_schema())
        pqwriter.write_table(chunk_table_filtered)
                
        n_processed += batch_size
                
pqwriter.close()
```
